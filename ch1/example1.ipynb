Pythonの基本的な文法の確認

```python
# 算術計算
print(1-2)
print(1+2)
print(4*5)
print(7/5)
print(3 ** 2)
```

```python
# データ型
print(type(10))
print(type(2.718))
print(type("hello"))
```

```python
# 変数
x = 10
print(x)
x = 20
print(x)
```

```python
# 型の自動変換
x = 100
y = 3.14
print(x * y)
print(type(x * y))
```

```python
# リスト
a = [1, 2, 3, 4, 5]
print(a)
print(len(a))
print(a[0])
a[4] = -5
print(a[4])
## スライシング
print(a[0:2])
print(a[2:])
print(a[:3])
print(a[:-1])
# エラー
print(a[5])
```

```python
# ディクショナリ
student = {'height':189}
print(student['height'])
student['weight'] = 70
print(student)
```

```python
# ブーリアン
hungry = True
sleepy = False
print(type(hungry))
print(not hungry)
print(hungry and sleepy)
print(hungry or sleepy)
```

```python
# if
hungry = True
if hungry:
    print("hungry!")
else:
    print("yum!")
```

```python
# for
for i in range(1,10,2):
    print(i)
```

```python
# function
def hello(name):
    print("Hello, " + name + "!")
hello("cat")
```

Pythonはインタプリタで対話的に動かす以外にもスクリプトとして保存して呼び出すこともできるよ。

```python
# class
class Man:
    def __init__(self, name):
        self.name = name
        print("init!")
    
    def hello(self):
        print("Hello" + self.name + "!")

m = Man("Tom")
m.hello()
```

Numpyの基本的な文法

```python
# import
import numpy as np
```

```python
x = np.array([1.0, 2.0, 3.0])
print(x)
print(type(x))
```

```python
x = np.array([1.0, 2.0, 3.0])
y = np.array([2.0, 4.0, 6.0])
# 要素毎の計算
print(x+y)
print(x-y)
print(x*y)
print(x/y)
# ブロードキャスト
print(x / 2.0)

2A
A * B
```

```python
# N次元配列
A = np.array([[1, 2], [3, 4]])
print(A)
print(A.shape)
print(A.dtype)
```

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[3, 0], [0, 6]])
print(A + B)
print(A * B)
# ブロードキャスト1
print(A * 10)
print(A * np.array([[10, 10], [10, 10]]))
# ブロードキャスト2
C = np.array([10, 20])
print(A * C)
print(A * np.array([[10, 20], [10, 20]]))
```

```python
# 要素へのアクセス
X = np.array([[1, 2], [3, 4]])
print(X)
print(X[0])
print(X[1, 1])
for row in X:
    print(row)
```

```python
# 1次元へ変換
X = np.array([[1, 2], [3, 4]])
print(X)
X = X.flatten()
print(X)
```

```python
# 要素の抜き出し
X = np.array([[1, 2], [3, 4]])
print(X)
print(X>1)
print(X[X>1])
```

Matplotlibの基本的な文法の確認

```python
# import
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline # plt.show()を省略できる
```

```python
x = np.arange(0, 6, 0.1)
y = np.sin(x)
# グラフの描画
plt.plot(x, y)
# plt.show() # 
```

```python
x = np.arange(0, 6, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)
# グラフの描画2
plt.plot(x, y1, label = "sin")
plt.plot(x, y2, label = "cos", linestyle = "--")
plt.xlabel("x")
plt.ylabel("y")
plt.title("sin vs cos")
plt.legend()
```
